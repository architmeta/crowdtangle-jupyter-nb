```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Search results from last 12 months (from July 2024) for keyword 'gaza' 
# gaza-pot.csv inlcude posts where the page admin is in Palestine Occupied Territories
# gaza-us inlcude posts where the page admin is in USA
# Page Categories filter: Media News Company, Newsagent Newsstand, News Personality, News Site, Newspaper, 
# Political Candidate, Political Party, Politician, Political Organisation 

df = pd.read_csv("gaza-pot.csv")
df.info()
```

```python
# Convert object-type columns to string, except integer columns
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str)
```

```python
unique_page_names = df['Page Name'].unique()
print(f"\nNumber of unique page names: {len(unique_page_names)}")
```

```python
# Get the unique page names
unique_page_names = df['Page Name'].unique()

# Print the results
print("Unique Page Names:")
for page_name in unique_page_names:
    print(f"Page Name: {page_name}")
    # Get the URL for the current page name
    url = df.loc[df['Page Name'] == page_name, 'URL'].values[0]
    print(f"URL: {url}")
    print()
```

```python
import csv

# Get the unique page names
unique_page_names = df['Page Name'].unique()

# Create a list to store the data
data = []

# Iterate through the unique page names
for page_name in unique_page_names:
    # Get the URL for the current page name
    url = df.loc[df['Page Name'] == page_name, 'URL'].values[0]
    # Get the Page Category for the current page name
    page_category = df.loc[df['Page Name'] == page_name, 'Page Category'].values[0]
    
    # Append the data to the list
    data.append({'Page Name': page_name, 'Page Category': page_category, 'URL': url})

# Write the data to a CSV file
with open('media_outlets_with_fb_page_admin_in_palestine.csv', 'w', newline='') as csvfile:
    fieldnames = ['Page Name', 'Page Category', 'URL']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for row in data:
        writer.writerow(row)
```

```python

# Convert object-type columns to string, except integer columns
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].astype(str)

# Define the keyword to search for
keyword = "genocide"

# Count the occurrences of the keyword in the "Message" column
keyword_counts = df[df['Message'].str.contains(keyword, case=False)]['Page Name'].value_counts()

# Print the keyword occurrences
print(f"The keyword '{keyword}' appears {keyword_counts.sum()} times in the 'Message' column.")
print(f"The breakdown by page is:\n{keyword_counts}")

# Get the top 5 most shared posts that mention the keyword
top_shared_posts = (
    df[df['Message'].str.contains(keyword, case=False)]
    .sort_values(by='Shares', ascending=False)
    .head(5)
    [['Page Name', 'URL', 'Shares']]
)

print("\nTop 5 most shared posts that mention the keyword:")
print(top_shared_posts)
```
